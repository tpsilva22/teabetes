Neste notebook vai ser feito uma análise exploratória a todas as variáveis numéricas.

Para cada variável vai ser feito os seguintes cálculos:

    mínimo
    máximo
    média
    mediana
    desvio padrão
    quartis

E os seguintes gráficos:

    Histograma 
    Boxplot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

diabetes = pd.read_csv("../data/diabetes_dataset.csv")
```

Variável: age

```python
diabetes["age"].describe()
```

```python
# Histograma
plt.figure(figsize=(12, 5))
sns.histplot(diabetes["age"], bins=72)
plt.title("Histogram - Age", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.show()

# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(y=diabetes["age"], width=0.5)
plt.title("Boxplot - Age", fontsize=16)
plt.ylabel("Age", fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.show()
```
