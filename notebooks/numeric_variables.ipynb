In this notebook, an exploratory analysis will be performed on all numerical variables.

For each variable, the following statistics will be calculated:

    minimum
    maximum
    mean
    median
    standard deviation
    quartiles
    asymmetry

And the following plots will be generated:

    Histogram
    Boxplot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

diabetes = pd.read_csv("../data/diabetes_dataset.csv")
```

Variable: age

```python
diabetes["age"].describe()
```

```python
diabetes["age"].skew()
```

```python
# Histogram
plt.figure(figsize=(12, 5))
sns.histplot(diabetes["age"], bins=72)
plt.title("Histogram - Age", fontsize=20)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.show()

# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(y=diabetes["age"], width=0.5)
plt.title("Boxplot - Age", fontsize=20)
plt.ylabel("Age", fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.show()
```
