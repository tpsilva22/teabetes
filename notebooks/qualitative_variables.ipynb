In this notebook, an exploratory analysis will be performed on all qualitative variables.

For each variable, the following statistics will be calculated:

    mode
    relative frequency
    absolute frequency

And the following plots will be generated:

    pie chart
    count plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt   
  
diabetes = pd.read_csv("../data/diabetes_dataset.csv")
```

variable: gender

```python
moda = diabetes["gender"].mode(dropna=True)
moda
```

```python
# Pie chart 
freq_rel = diabetes["gender"].value_counts(normalize=True, dropna=False) * 100
plt.pie(freq_rel, labels=freq_rel.index, autopct='%1.2f%%')
plt.title('Relative Frequency - Gender')
plt.show()
```

```python
# Count plot 
diag_label = diabetes["diagnosed_diabetes"].map({0: "Não", 1: "Sim"})
sns.countplot(data=diabetes, x="gender", hue=diag_label)
plt.title('Absolute Frequency - Gender')
plt.legend(title="Diagnosed Diabetes")
freq_abs=diabetes["gender"].value_counts(dropna=False)
# Shows the values represented in the chart (absolute frequency) in numeric form
print(freq_abs)
```

variable: diabetes_stage

```python
moda = diabetes ["diabetes_stage"].mode(dropna=True) 
moda      
```

```python
# Pie chart 
freq_rel = diabetes["diabetes_stage"].value_counts(normalize=True, dropna=False) * 100
plt.pie(freq_rel, startangle=140)
# Legend with labels and percentages
legend_labels = [f'{cat}: {val:.2f}%' for cat, val in freq_rel.items()]
plt.legend(legend_labels, loc='center left', bbox_to_anchor=(1, 0, 0.5, 1), fontsize=11)

plt.title('Relative Frequency - Diabetes Stage')
plt.show()
```

```python
# Count plot 
sns.countplot(data=diabetes, x="diabetes_stage")
plt.title('Absolute Frequency - Diabetes Stage')
freq_abs=diabetes["diabetes_stage"].value_counts(dropna=False)
# Shows the values represented in the chart (absolute frequency) in numeric form
print(freq_abs)
```
