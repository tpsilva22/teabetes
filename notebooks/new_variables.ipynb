In this notebook, we will create new variables such as:

    Insulin_Resistance
    Glycemia_Spike
    Obesity_Index
    Pulse_Pressure
    MAP (Mean Arterial Pressure)
    Ratio_LDL_HDL
    Clinical_Risk_Score
    Age_Groups
    Weight_Status

After creating them, we will generate a new dataset that will be used for analysis in the other notebooks.

```python
import pandas as pd

diabetes = pd.read_csv("../data/diabetes_dataset.csv")

diabetes["insulin_resistance"] = (diabetes["glucose_fasting"] * diabetes["insulin_level"])/405
diabetes["glycemia_spike"] = diabetes["glucose_postprandial"] - diabetes["glucose_fasting"]
diabetes["obesity_index"] = diabetes["bmi"] * diabetes["waist_to_hip_ratio"]
diabetes["pulse_pressure"] = diabetes["systolic_bp"] - diabetes["diastolic_bp"]
diabetes["map"] = diabetes["diastolic_bp"] + 1/3 * (diabetes["systolic_bp"] - diabetes["diastolic_bp"])
diabetes["ratio_ldl_hdl"] = diabetes["ldl_cholesterol"] / diabetes["hdl_cholesterol"]
diabetes["clinical_risk_score"] = diabetes["hypertension_history"] + diabetes["cardiovascular_history"] + diabetes["family_history_diabetes"] + diabetes["family_history_diabetes"] 
diabetes["age_groups"] = pd.cut(diabetes["age"], bins=[0, 35, 50, 65, 120], labels=['Young Adult', 'Adult', 'Senior Adult', 'Elderly'])
diabetes["weight_status"] = pd.cut(diabetes["bmi"], bins=[0, 18.5, 24.9, 29.9, 100], labels=['Underweight', 'Normal', 'Overweight', 'Obese'])

diabetes.to_csv("../data/diabetes_dataset_new_variables.csv", index=False)
```
