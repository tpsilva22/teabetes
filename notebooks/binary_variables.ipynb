In this notebook, an exploratory analysis will be performed on all binary variables.

For each variable, the following statistics will be calculated:

    absolute frequency
    relative frequency

And the following plots will be generated:

    pie chart
    count plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

diabetes = pd.read_csv("../data/diabetes_dataset_new_variables.csv")
```

variable: family_history_diabetes

```python
# Pie chart
family_history_label = diabetes["family_history_diabetes"].replace({0: "No", 1: "Yes"})
freq_rel = family_history_label.value_counts(normalize=True, dropna=False) * 100
plt.pie(freq_rel, labels=freq_rel.index, autopct='%1.2f%%')
plt.title('Relative Frequency - Family History of Diabetes')
plt.show()
```

```python
# Count plot
family_history_label = diabetes["family_history_diabetes"].replace({0: "No", 1: "Yes"})
diag_label = diabetes["diagnosed_diabetes"].replace({0: "No", 1: "Yes"})
sns.countplot(x=family_history_label, hue=diag_label)
plt.title('Absolute Frequency - Family History of Diabetes')
plt.xlabel('Family History of Diabetes')
plt.legend(title="Diagnosed Diabetes")
freq_abs = family_history_label.value_counts(dropna=False)
# Shows the values represented in the chart (absolute frequency) in numeric form
print(freq_abs)
```
