# Dataset Description

The following table describes the variables present in the dataset used in this project, including each data type and what each attribute represents.

This dataset belongs to the **healthcare** domain, with a focus on clinical, metabolic, and behavioral factors associated with diabetes risk and diagnosis.

The variables that include "(New)" are newly engineered features added to the dataset.

| Variable | Type | What it Represents |
|---|---|---|
| age | Numeric | Individual's chronological age. |
| gender | Categorical | Biological or self-identified gender. |
| ethnicity | Categorical | Ethnic and cultural background. |
| education_level | Categorical | Highest educational qualification attained. |
| income_level | Categorical | Annual household income bracket. |
| employment_status | Categorical | Current work or professional situation. |
| smoking_status | Categorical | History and current intensity of tobacco use. |
| alcohol_consumption_per_week | Numeric | Volume of alcohol intake on a weekly basis. |
| physical_activity_minutes_per_week | Numeric | Total duration of physical exercise performed weekly. |
| diet_score | Numeric | Numerical assessment of nutritional habits and food quality. |
| sleep_hours_per_day | Numeric | Average daily duration of restorative sleep. |
| screen_time_hours_per_day | Numeric | Daily time dedicated to digital device usage. |
| family_history_diabetes | Binary | Genetic predisposition based on close relatives' diagnoses. |
| hypertension_history | Binary | Clinical record of high blood pressure diagnosis. |
| cardiovascular_history | Binary | Medical history of heart-related conditions or strokes. |
| bmi | Numeric | Body Mass Index; a general measure of body fat based on height and weight. |
| waist_to_hip_ratio | Numeric | Measure of fat distribution and central (visceral) obesity. |
| systolic_bp | Numeric | Peak arterial pressure during a heartbeat (mmHg). |
| diastolic_bp | Numeric | Minimum arterial pressure between heartbeats (mmHg). |
| heart_rate | Numeric | Number of heart contractions per minute at rest. |
| cholesterol_total | Numeric | Overall concentration of cholesterol in the bloodstream. |
| hdl_cholesterol | Numeric | High-Density Lipoprotein ("good" cholesterol) levels. |
| ldl_cholesterol | Numeric | Low-Density Lipoprotein ("bad" cholesterol) levels. |
| triglycerides | Numeric | Concentration of the most common type of fat found in the blood. |
| glucose_fasting | Numeric | Blood sugar levels measured after a period without food intake. |
| insulin_level | Numeric | Amount of insulin hormone circulating in the blood. |
| hba1c | Numeric | Long-term average of blood sugar levels (last 2-3 months). |
| diabetes_stage | Categorical | Severity and clinical categorization of the diabetic condition. |
| diagnosed_diabetes | Binary | Target Variable: Confirmation of a clinical diabetes diagnosis. |
| diabetes_risk_score | Numeric | Composite risk estimate based on lifestyle and health factors. |
| glucose_postprandial | Numeric | Blood sugar levels measured specifically after meal consumption. |
| insulin_resistance | Numeric | (New) Assessment of how poorly the body responds to insulin (HOMA-IR). |
| glycemia_spike | Numeric | (New) Measure of the body's glucose tolerance after a meal. |
| obesity_index | Numeric | (New) Refined metabolic risk indicator combining mass and fat location. |
| pulse_pressure | Numeric | (New) Indicator of arterial stiffness and cardiovascular stress. |
| map | Numeric | (New) Average blood pressure required to perfuse vital organs. |
| ratio_ldl_hdl | Numeric | (New) Relative balance between "bad" and "good" cholesterol (heart risk). |
| clinical_risk_score | Numeric | (New) Cumulative burden of genetic and clinical history factors. |
| age_groups | Categorical | (New) Segmentation of the sample into distinct life stages. |
| weight_status | Categorical | (New) Clinical classification of body weight (Underweight to Obese). |
