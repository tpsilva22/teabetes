In this notebook, exploratory analysis will be performed based on correlations.

First, we will create three heatmaps using three correlation coefficients, one for each heatmap.

Next, we will create plots for critical, strong, moderate, and weak correlations based on our heatmaps.

For numeric-numeric pairs, we will use scatter plots, and for numeric-binary pairs, we will use boxplots.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

diabetes = pd.read_csv("../data/diabetes_dataset.csv")

# Numeric variables 
numeric_vars = [
    "age",
    "bmi",
    "waist_to_hip_ratio",
    "diet_score",
    "sleep_hours_per_day",
    "screen_time_hours_per_day",
    "alcohol_consumption_per_week",
    "physical_activity_minutes_per_week",
    "systolic_bp",
    "diastolic_bp",
    "heart_rate",
    "cholesterol_total",
    "hdl_cholesterol",
    "ldl_cholesterol",
    "triglycerides",
    "glucose_fasting",
    "insulin_level",
    "hba1c",
    "diabetes_risk_score",
    "glucose_postprandial",
]

# Binary variables
binary_vars = [
    "family_history_diabetes",
    "hypertension_history",
    "cardiovascular_history",
    "diagnosed_diabetes"
]

all_vars = numeric_vars + binary_vars

corr_matrix_pearson = diabetes[numeric_vars].corr(method="pearson")
corr_matrix_spearman = diabetes[all_vars].corr(method="spearman")
corr_matrix_kendall = diabetes[all_vars].corr(method="kendall")
```

```python
# Pearson Correlation Heatmap (20 Numeric Variables Only)

plt.figure(figsize=(14, 12))
sns.heatmap(
    corr_matrix_pearson,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    linewidths=0.5
)
plt.title("Pearson Correlation - Numeric Variables", fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Spearman Correlation Heatmap (20 Numeric + 4 Binary Variables)

plt.figure(figsize=(16, 14))
sns.heatmap(
    corr_matrix_spearman,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    linewidths=0.5
)
plt.title("Spearman Correlation - All Variables", fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Kendall Correlation Heatmap (20 Numeric + 4 Binary Variables)

plt.figure(figsize=(16, 14))
sns.heatmap(
    corr_matrix_kendall,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    linewidths=0.5
)
plt.title("Kendall Correlation - All Variables\n", fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

Critical Correlations

```python
# HbA1c vs Glucose Postprandial
plt.figure(figsize=(12, 5))
plt.scatter(diabetes["hba1c"], diabetes["glucose_postprandial"], alpha=0.5)
plt.title("Glycated Hemoglobin vs Glucose Postprandial")
plt.xlabel("Glycated Hemoglobin")
plt.ylabel("Glucose Postprandial")
plt.tight_layout()
plt.show()
```

```python
# Cholesterol Total vs LDL Cholesterol
plt.figure(figsize=(12, 5))
plt.scatter(diabetes["cholesterol_total"], diabetes["ldl_cholesterol"], alpha=0.5)
plt.title("Cholesterol Total vs LDL Cholesterol")
plt.xlabel("Cholesterol Total")
plt.ylabel("LDL Cholesterol")
plt.tight_layout()
plt.show()
```

```python
# BMI vs Waist-to-Hip Ratio
plt.figure(figsize=(12, 5))
plt.scatter(diabetes["bmi"], diabetes["waist_to_hip_ratio"], alpha=0.5)
plt.title("BMI vs Waist-to-Hip Ratio")
plt.xlabel("BMI")
plt.ylabel("Waist-to-Hip Ratio")
plt.tight_layout()
plt.show()
```

```python
# HbA1c vs Diagnosed Diabetes (Boxplot)
plt.figure(figsize=(12, 5))
sns.boxplot(data=diabetes, x="diagnosed_diabetes", y="hba1c")
plt.title("Glycated Hemoglobin by Diagnosed Diabetes")
plt.xlabel("Diagnosed Diabetes")
plt.ylabel("Glycated Hemoglobin")
plt.show()
```
