```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

diabetes = pd.read_csv("../data/diabetes_dataset.csv")

numeric_vars = [
    "age",
    "bmi",
    "waist_to_hip_ratio",
    "diet_score",
    "sleep_hours_per_day",
    "screen_time_hours_per_day",
    "alcohol_consumption_per_week",
    "physical_activity_minutes_per_week",
    "systolic_bp",
    "diastolic_bp",
    "heart_rate",
    "cholesterol_total",
    "hdl_cholesterol",
    "ldl_cholesterol",
    "triglycerides",
    "glucose_fasting",
    "insulin_level",
    "hba1c",
    "diabetes_risk_score",
    "glucose_postprandial",
]

corr_matrix_pearson = diabetes[numeric_vars].corr(method="pearson")
corr_matrix_spearman = diabetes[numeric_vars].corr(method="spearman")
corr_matrix_kendall = diabetes[numeric_vars].corr(method="kendall")
```

```python
# Pearson Correlation Heatmap

plt.figure(figsize=(16, 12))

sns.heatmap(
    corr_matrix_pearson,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    linewidths=0.5
)

plt.title("Pearson Correlation Heatmap - Numeric Variables\n", fontsize=16)
plt.xticks(rotation=90, ha="left")
plt.yticks(rotation=0)
plt.show()
```

```python
# Spearman Correlation Heatmap

plt.figure(figsize=(16, 12))

sns.heatmap(
    corr_matrix_spearman,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    linewidths=0.5
)

plt.title("Spearman Correlation Heatmap - Numeric Variables\n", fontsize=16)
plt.xticks(rotation=90, ha="left")
plt.yticks(rotation=0)
plt.show()
```

```python
# Kendall Correlation Heatmap

plt.figure(figsize=(16, 12))
sns.heatmap(
    corr_matrix_kendall,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    linewidths=0.5
)
plt.title("Kendall Correlation Heatmap - Numeric Variables\n", fontsize=16)
plt.xticks(rotation=90, ha="left")
plt.yticks(rotation=0)
plt.show()
```
