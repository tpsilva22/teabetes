```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler  
```

```python
diabetes = pd.read_csv("../data/diabetes_dataset_new_variables.csv")
cols_to_scale = [
    "age",
    "alcohol_consumption_per_week",
    "physical_activity_minutes_per_week",
    "diet_score",
    "sleep_hours_per_day",
    "screen_time_hours_per_day",
    "bmi",
    "waist_to_hip_ratio",
    "systolic_bp",
    "diastolic_bp",
    "heart_rate",
    "cholesterol_total",
    "hdl_cholesterol",
    "ldl_cholesterol",
    "triglycerides",
    "glucose_fasting",
    "insulin_level",
    "hba1c",
    "diabetes_risk_score",
    "glucose_postprandial",
    "insulin_resistance",
    "glycemia_spike",
    "obesity_index",
    "pulse_pressure",
    "map",
    "ratio_ldl_hdl",
    "clinical_risk_score"
]

diabetes_minmax = diabetes.copy()
diabetes_standard = diabetes.copy()

missing=diabetes.loc[diabetes.isna().any(axis=1)]
missing

# Does not exist null values in the dataset
```

```python
# Normalization (Min-Max) -> Everything between 0 and 1
scaler_minmax = MinMaxScaler()
diabetes_minmax[cols_to_scale] = scaler_minmax.fit_transform(diabetes[cols_to_scale])

# Standardization (StandardScaler) -> Mean 0, Standard Deviation 1
scaler_standard = StandardScaler()
diabetes_standard[cols_to_scale] = scaler_standard.fit_transform(diabetes[cols_to_scale])
```

In the cell below, we compare the data of one column before standardization with the data after standardization

```python
age_before_mean = diabetes['age'].mean()
age_before_std = diabetes['age'].std()

age_after_mean = diabetes_standard['age'].mean()
age_after_std = diabetes_standard['age'].std()

print(f"Age antes da standardization -> mean: {age_before_mean:.4f}, std: {age_before_std:.4f}")
print(f"Age depois da standardization -> mean: {age_after_mean:.4f}, std: {age_after_std:.4f}")
```

In the cell below, we compare the data of one column before normalization with the data after normalization 

```python
age_before_min = diabetes['age'].min()
age_before_max = diabetes['age'].max()

age_after_min = diabetes_minmax['age'].min()
age_after_max = diabetes_minmax['age'].max()

print(f"Age antes da normalizacao -> min: {age_before_min}, max: {age_before_max}")
print(f"Age depois da normalizacao -> min: {age_after_min}, max: {age_after_max}")
```
