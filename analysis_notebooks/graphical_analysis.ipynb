Imports


```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

Dataset


```python
pd.set_option('display.max_columns', None)
data = pd.read_csv("../data/diabetes_dataset_new_variables.csv")
data
```

1. Risk Analysis


```python
sns.boxplot(x='age_groups', y='diabetes_risk_score', data=data, order=['Young Adult', 'Adult', 'Senior Adult', 'Elderly'])
plt.title("Risk vs Age Group")
plt.show()
```

Increasing age is strongly associated with a higher risk of diabetes.

```python
sns.boxplot(x='family_history_diabetes', y='diabetes_risk_score', data=data)
plt.title("Diabetes Risk Score vs Family History")
plt.show()
```

A family history of diabetes greatly influences a person's risk of developing the disease.

```python
sns.scatterplot(x='diabetes_risk_score', y='insulin_resistance', hue='weight_status', data=data)
sns.regplot(data=data, x="diabetes_risk_score", y="insulin_resistance", scatter=False)
plt.title("Diabetes Risk Score vs Insulin resistance")
plt.show()

sns.scatterplot(x='diabetes_risk_score', y='waist_to_hip_ratio', hue='weight_status', data=data)
sns.regplot(data=data, x="diabetes_risk_score", y="waist_to_hip_ratio", scatter=False)
plt.title("Diabetes Risk Score vs Waist to Hip Ratio")
plt.show()
```

Insulin resistance is a key factor associated with increased diabetes risk. While obesity is linked to higher insulin resistance, it is not a sole determinant, as factors such as fat distribution play an important role.

```python
sns.scatterplot(x='physical_activity_minutes_per_week', y='insulin_resistance', data=data)
sns.regplot(data=data, x="physical_activity_minutes_per_week", y="insulin_resistance", scatter=False)
plt.title("Insulin resistance vs Physical activity per Week")
plt.show()

sns.scatterplot(x='physical_activity_minutes_per_week', y='glucose_fasting', data=data)
sns.regplot(data=data, x="physical_activity_minutes_per_week", y="glucose_fasting", scatter=False)
plt.title("Glucose Fasting vs Physical activity per Week")
plt.show()

sns.scatterplot(x='physical_activity_minutes_per_week',y='diabetes_risk_score', data=data)
sns.regplot(data=data, x="physical_activity_minutes_per_week", y="diabetes_risk_score", scatter=False)
plt.title("Diabetes Risk Score vs Physical Activity per Week")
plt.show
```

Physical activity reduces blood glucose levels and helps decrease insulin resistance, which is important for reducing the risk of diabetes.

```python
sns.scatterplot(x='diet_score',y='bmi', data=data)
sns.regplot(data=data, x="diet_score", y="bmi", scatter=False)
plt.title("Diet Score vs BMI")
plt.show()

sns.scatterplot(x='diabetes_risk_score',y='bmi', data=data)
sns.regplot(data=data, x="diabetes_risk_score", y="bmi", scatter=False)
plt.title("Diabetes Risk Score vs BMI")
plt.show()

sns.scatterplot(x='diabetes_risk_score',y='diet_score', data=data)
sns.regplot(data=data, x="diabetes_risk_score", y="diet_score", scatter=False)
plt.title("Diabetes Risk Score vs Diet Score")
plt.show()
```

Higher diet scores are associated with lower BMI and will contribute to reduced diabetes risk.

2. Diagnostic Analysis

```python
sns.boxplot(x='diagnosed_diabetes', y='glucose_fasting', data=data)
plt.title("Glucose Fasting vs Diabetes")
plt.show()
```

Fasting blood glucose levels above 120-130 confirm that the patient has diabetes, but it does not mean that all diabetics have high blood glucose levels.

```python
sns.scatterplot(x='insulin_resistance', y='insulin_level', hue='diagnosed_diabetes', data=data)
plt.title("Insulin Level vs Insulin Resistance")
plt.show()
```

The greater the resistance, the more insulin the body must produce to compensate; if the increase is not significant, the patient is likely to have diabetes.

```python
sns.boxplot(x='diagnosed_diabetes', y='hba1c', data=data)
plt.title("Hba1c vs Diagnosed Diabetes")
plt.show()
```

Individuals with HbA1c (Glycated Hemoglobin) values ​​above the range of 6.5-6.8 are diagnosed with diabetes.

3. Diabetes type Analysis.

```python
sns.boxplot(x='weight_status', y='insulin_resistance', data=data, order=['Normal', 'Obese', 'Overweight', 'Underweight'])
plt.title("Weight Status vs Insulin Resistance")
plt.show()

#Normalizar dados
prop = pd.crosstab(
    data['weight_status'],
    data['diabetes_stage'],
    normalize='index'   
)

prop.plot(kind='bar', stacked=True)
plt.title("Weight Status vs Diabetes Stage")
plt.xticks(rotation=45)
plt.ylabel("Proportion")
plt.legend(title='Diabetes Stage')
plt.show()
```

Fat increases insulin resistance, making type 2 diabetes more common in the overweight/obese group of people.

```python
counts = data["diabetes_stage"].value_counts()

plt.figure()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title("Distribution of Diabetes Stages")
plt.show()
```

The dataset is not suitable for studying type 1 and Gestational diabetes values, ​​due to the small amount of related data.

```python
data_filtered = data[~data["diabetes_stage"].isin(["Type 1", "Gestational"])]

sns.boxplot(x='diabetes_stage', y='glucose_fasting', data=data_filtered)
plt.title("Diabetes Stage vs Glucose Fasting")
plt.show()

sns.boxplot(x='diabetes_stage', y='glucose_postprandial', data=data_filtered)
plt.title("Diabetes Stage vs Glucose Postprandial")
plt.show()

sns.boxplot(x='diabetes_stage', y='hba1c', data=data_filtered)
plt.title("Diabetes Stage vs Hba1c")
plt.show()
```

Fasting glucose levels above the 98-100 range, or HbA1c values ​​above 5.8, indicates that the patient is at least pre-diabetic.

4. Behavior Analysis

```python
sns.boxplot(x='age_groups', y='screen_time_hours_per_day', data=data)
plt.title("Screen Time Hours per Day vs Age Group")
plt.show()

sns.boxplot(x='age_groups', y='physical_activity_minutes_per_week', data=data)
plt.title("Physical activity minutes per Week vs Age Group")
plt.show()

data.groupby('age_groups')[
    ['screen_time_hours_per_day', 'physical_activity_minutes_per_week']
].mean()
```

The dataset is not suitable for studying age-related behaviors.

```python
counts = data["smoking_status"].value_counts()

plt.figure()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title("Distribution of Smoking Status")
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x="smoking_status", y="diabetes_risk_score", data=data)
plt.title("Diabetes Risk by Smoking Status")

plt.show()
```

Smoking status is not related to the increase of diabetes risk in this dataset.

5. Metabolic Analysis

```python
sns.scatterplot(x='age', y='systolic_bp', data=data)
sns.regplot(data=data, x="age", y="systolic_bp", scatter=False)
plt.title("Insulin Resistance vs Systolic Bp")
plt.show()

sns.scatterplot(x='age', y='pulse_pressure', data=data)
sns.regplot(data=data, x="age", y="pulse_pressure", scatter=False)
plt.title("Age vs Pulse Pressure")
plt.show()

sns.scatterplot(x='insulin_resistance', y='systolic_bp', hue="age_groups", data=data)
sns.regplot(data=data, x="insulin_resistance", y="systolic_bp", scatter=False)
plt.title("Insulin Resistance vs Systolic Bp")
plt.show()
```

Systolic blood pressure (SBP) tends to increase with age and is related to increased insulin resistance, causing the risk of diabetes to increase with age.

```python

sns.scatterplot(data=data, x="bmi", y="triglycerides", hue="diagnosed_diabetes")
sns.regplot(data=data, x="bmi", y="triglycerides", scatter=False)
plt.title("Triglicerídeos vs BMI")
plt.show()

sns.scatterplot(data=data, x="insulin_resistance", y="triglycerides", hue="diagnosed_diabetes")
sns.regplot(data=data, x="insulin_resistance", y="triglycerides", scatter=False)
plt.title("Triglicerídeos vs Insulin Resistance")
plt.show()

sns.scatterplot(data=data, x="insulin_resistance", y="ldl_cholesterol", hue="diagnosed_diabetes")
sns.regplot(data=data, x="insulin_resistance", y="ldl_cholesterol", scatter=False)
plt.title("LDL Cholesterol vs Insulin Resistance")
plt.show()

```

Triglycerides and cholesterol do not directly cause diabetes, but they are involved in the same metabolic process and may indirectly contribute to its development by increasing insulin resistance.
