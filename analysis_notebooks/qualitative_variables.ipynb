In this notebook, an exploratory analysis will be performed on all qualitative variables.

For each variable, the following statistics will be calculated:

    mode
    relative frequency
    absolute frequency

And the following plots will be generated:

    pie chart
    count plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt   
  
diabetes = pd.read_csv("../data/diabetes_dataset_new_variables.csv")
```

variable: gender

```python
moda = diabetes["gender"].mode(dropna=True)
moda
```

```python
# Pie chart 
freq_rel = diabetes["gender"].value_counts(normalize=True, dropna=False) * 100
plt.pie(freq_rel, labels=freq_rel.index, autopct='%1.2f%%')
plt.title('Relative Frequency - Gender')
plt.show()
```

```python
# Count plot 
diag_label = diabetes["diagnosed_diabetes"].map({0: "No", 1: "Yes"})
sns.countplot(data=diabetes, x="gender", hue=diag_label)
plt.title('Absolute Frequency - Gender')
plt.legend(title="Diagnosed Diabetes")
freq_abs=diabetes["gender"].value_counts(dropna=False)
# Shows the values represented in the chart (absolute frequency) in numeric form
print(freq_abs)
```

Gender appears relatively balanced in the sample, which reduces sex-related sampling bias. The split by diagnosed diabetes suggests that risk is present across both groups, indicating that gender alone is not a sufficient explanatory factor and should be interpreted together with metabolic and lifestyle variables.
